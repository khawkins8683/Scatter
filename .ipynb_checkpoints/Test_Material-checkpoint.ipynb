```python
import numpy as np
import matplotlib.pyplot as plt

from Code import material as mat
```

### Phase Functions

```python
### Iso troptic
[thetas, phis] = mat.IsotropicPhaseFunction(1000)
plt.scatter(np.arange(0,len(thetas),1),thetas)
plt.scatter(np.arange(0,len(phis),1),phis)
```

```python
#Henyey-Greenstein
g=-0.35
thetaList = np.linspace(0,np.pi,200)
probList = mat.HG(thetaList,g)
plt.plot(thetaList,probList)
plt.ylim(0,0.08)
```

```python
testtheta = []
testphi = []
for i in range(10000):
    v= mat.HGPhaseFunction(g)
    testtheta.append(v[0])
    testphi.append(v[1])
plt.hist(np.array(testtheta)[:],20)
plt.show()
plt.hist(np.array(testphi)[:],20)
```

### Material Class

```python
air= mat.Material(1.5+0.0001*1j, 1000 , 1E-5)
print(air)
```

```python
diffuseMat = mat.HGMaterial(1.5+0.0001*1j, 1E-5 ,1E-5,0.3)
print(diffuseMat)
```

```python
testtheta = []
testphi = []
for i in range(10000):
    v= air.phaseFunction({})
    testtheta.append(v[0])
    testphi.append(v[1])
plt.hist(np.array(testtheta)[:],20)
plt.show()
plt.hist(np.array(testphi)[:],20)
```

```python
testtheta = []
testphi = []
for i in range(10000):
    v= diffuseMat.phaseFunction({})
    testtheta.append(v[0])
    testphi.append(v[1])
plt.hist(np.array(testtheta)[:],20)
plt.show()
plt.hist(np.array(testphi)[:],20)
```
